```python
import os

root_folder = "k_fold_tune"
k_fold_tunes = [f for f in os.listdir(root_folder) 
        if os.path.isdir(os.path.join(root_folder, f)) 
            and not f.endswith(".ipynb_checkpoints")]
```

```python
import numpy as np
import json
import pandas as pd
from collections import defaultdict

from data_utils import get_seq_list
from train_utils import default_args, retrain
from eval_utils import get_all_dfs, print_all_metrics

eval_dict = defaultdict(list)

for k_fold_tune_idx, k_fold_tune_name in enumerate(k_fold_tunes):
    k_fold_tune_path = os.path.join(root_folder, k_fold_tune_name)
    print(k_fold_tune_path)
    tuning_names = [f for f in os.listdir(k_fold_tune_path) if not f.endswith("checkpoints")]
    tuning_names = sorted(tuning_names)
    print("Num tunings: ", len(tuning_names))
    # get default args
    args, model_args = default_args()
    # load tuning args into default args
    first_tuning_path = os.path.join(k_fold_tune_path, tuning_names[0]) # to get base args
    with open(os.path.join(first_tuning_path, "args.json"), "r") as f:
        tuning_args = json.load(f)
    for key in tuning_args:
        args[key] = tuning_args[key]
    # load data
    seq_list = get_seq_list(args["minutes"], False, "ICP_Vital", args["features"], verbose=False)
    seq_list = np.array(seq_list)
    for tuning_idx, tuning_name in enumerate(tuning_names):
        tuning_path = os.path.join(k_fold_tune_path, tuning_name)
        print(tuning_path)
        # load test idcs and make dev/test split for this tuning run
        test_idcs = np.load(os.path.join(tuning_path, "test_idcs.npy"),
                           allow_pickle=True)
        # in some tunings I saved all test_idcs by accident for each tuning
        if isinstance(test_idcs.dtype, object):
            test_idcs = test_idcs[tuning_idx]
        test_idcs = test_idcs.astype(int)
        dev_idcs = np.array([idx for idx in range(len(seq_list)) if idx not in test_idcs]).astype(int)
        dev_data, test_data = seq_list[test_idcs], seq_list[dev_idcs]
        # load best params
        with open(os.path.join(first_tuning_path, "best_params.json"), "r") as f:
            best_params = json.load(f)
        # retrain
        models, trainers = retrain(dev_data, test_data, best_params, args, model_args, verbose=False)
        # eval on test
        pred_df = get_all_dfs(models, trainers, 
                         args["model_type"], 
                         dl_type="test")
        mean_test_error = pred_df["error"].mean()
        print("Mean test error: ", mean_test_error)
        # accumulate metrics
        eval_dict["test_error"].append(mean_test_error)
        eval_dict["k_fold_tune"].append(k_fold_tune_idx)
        eval_dict["tuning"].append(tuning_idx)
```
